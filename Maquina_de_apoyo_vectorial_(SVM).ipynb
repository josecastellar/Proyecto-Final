```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
final_df = pd.read_excel('FGR_data_para_proyecto_final_clean.xlsx')
print(final_df.shape)
final_df.head(100)
```

```python
final_df.isnull().sum()
```

```python
final_df.info()
```

```python
X = final_df.drop('Fetal weight', axis=1)
y = final_df['Fetal weight']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
rs_param_grid = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [0.1, 1, 10, 100, 300],
    'gamma': [0.00001, 0.0001, 0.001, 0.01, 0.1, 1]
}
```

```python
clf = svm.SVC()
```

```python
clf_rs = RandomizedSearchCV(
    estimator=clf,
    param_distributions=rs_param_grid,
    cv=3,
    verbose=2,
)

clf_rs.fit(X_train, y_train)




  #n_iter=10000,
```

```python
print('Los mejores parametros encontrados son: ',clf_rs.best_params_)
print('La mejor exactitud encontrada fue: ',clf_rs.best_score_)
```

```python
best_model = clf_rs.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Exactitud de :', accuracy)
```
